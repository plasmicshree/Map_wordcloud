```python
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt 
import pandas as pd 
from PIL import Image
```

```python
df=pd.read_csv('abstract.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
stopwords=set(STOPWORDS)
new_stop=['Wa','using','show','time','result','different',
          'of','effect of','study of','properties of','due to','study of']
en_stop=stopwords.union(new_stop)
comment_words = ''
en_stop
```

```python
import numpy as np
nepal=np.array(Image.open('np.png'))
flag=np.array(Image.open('flagy.png'))
nepal
```

```python

for val in df['Abstract']: 
      
    # typecaste each val to string 
    val = str(val) 
  
    # split the value 
    tokens = val.split() 
      
    # Converts each token into lowercase 
    for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower() 
      
    comment_words += " ".join(tokens)+" "
  
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = en_stop,mask=nepal, 
                min_font_size = 10).generate(comment_words) 
  
# plot the WordCloud image                        
plt.figure(figsize = (12, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show()
```

```python
def random_color_func(word=None, font_size=None, position=None,  orientation=None, font_path=None, random_state=None):
    h = int(255.0 * 21.0 / 255.0)
    s = int(250.0 * 255.0 / 255.0)
    l = int(100.0 * float(random_state.randint(60, 120)) / 255.0)

    return "hsl({}, {}%, {}%)".format(h, s, l)
```

```python
for val in df['Title of the Presentation']: 
      
    # typecaste each val to string 
    val = str(val) 
  
    # split the value 
    tokens = val.split() 
      
    # Converts each token into lowercase 
    for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower() 
      
    comment_words += " ".join(tokens)+" "
  
wordcloud3 = WordCloud(width = 800, height = 800, 
                background_color ='white', mask=flag,
                stopwords = en_stop,color_func=random_color_func,
                       contour_width=6, contour_color='blue', 
                min_font_size = 10).generate(comment_words) 
  
# plot the WordCloud image                        
plt.figure(figsize = (8, 10), facecolor = None) 
plt.imshow(wordcloud3) 
plt.axis("off") 
plt.tight_layout(pad = 1) 
plt.show()
```
